```python
import sys
from pathlib import Path

ROOT = Path().resolve().parent.parent
sys.path.append(str(ROOT))
```

```python
image = 'eso0930a'
sim = 'sim1'

data_folder = ROOT / 'data'

ground_truth = data_folder / 'processed' / image / 'groundtruth.fits'
image_folder = data_folder / 'processed' / image / sim
```

```python
from src.preprocess.image_quality_metrics import ImageQualityMetrics

print("Hogbom reconstruction quality metrics:")
image_path = image_folder/"tclean_hogbom_50000"/'tclean_hogbom_50000.model'

iqm = ImageQualityMetrics(
    image_path=str(image_path),
    reference_path=str(ground_truth)
)

psnr = iqm.psnr_reference()
ssim = iqm.ssim()


print(f"PSNR: {psnr:.2f}, SSIM: {ssim:.4f}\n")

iqm.plot_images()
```
