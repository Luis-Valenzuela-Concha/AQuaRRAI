```python
import sys
from pathlib import Path
import os 
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path().resolve().parent.parent

sys.path.append(str(ROOT))
```

```python
file = ROOT / "data/id_raw.csv"

df = pd.read_csv(file, sep=',')
```

```python
object_types = df['source_type'].value_counts()

plt.figure(figsize=(12, 6))
object_types.plot(kind='barh', color='#013966', zorder=2)
plt.title('Distribution of Object Types')
plt.xlabel('Count')
plt.ylabel('Object Type')
plt.grid(axis='x', linestyle='-', alpha=0.7, zorder=1)
plt.show()
```

```python
df_decals = df[df['source'] == 'decals']
df_eso = df[df['source'] == 'eso']
```

```python
eso_example_path = ROOT / "data/raw/eso/eso0332a.jpg"
eso_example = plt.imread(eso_example_path)

decals_example_path = ROOT / "data/raw/galaxy10_decals/image_116_label_5.png"
decals_example = plt.imread(decals_example_path)

fig, ax = plt.subplots(1, 2, figsize=(8, 8))
ax[0].imshow(eso_example)
ax[0].set_title('Example from ESO Dataset')
ax[0].axis('off')
ax[1].imshow(decals_example)
ax[1].set_title('Example from Decals Dataset')
ax[1].axis('off')
plt.show()
```
